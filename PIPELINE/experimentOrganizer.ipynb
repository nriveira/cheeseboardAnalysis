```python
from experimentStruct import ExperimentStruct
import matplotlib.pyplot as plt
import os
```

```python
# Load all files in a folder and subfolders that end with '_timestamps.csv' and create an ExperimentStruct for each file
timestamp_folder = r'/Users/nick/Projects/cheeseboardAnalysis/DATA/NOVEMBER/COMBINED/'

experiments = []
for root, dirs, files in os.walk(timestamp_folder):
    for file in files:
        if file.endswith('_timestamps.csv'):
            timestamp_file = os.path.join(root, file)
            experiment = ExperimentStruct(timestamp_file)
            experiments.append(experiment)

# Sort experiments by datetime
experiments.sort(key=lambda x: x.experimentDatetime)
# Print sorted experiment datetimes
for exp in experiments:
    print(f"Experiment Datetime: {exp.experimentDatetime}, Tag: {exp.experimentTag}")

# Add all file locations to a yaml file to be organized into trial epochs for comparison across days and trial types
yaml_file = r'/Users/nick/Projects/cheeseboardAnalysis/DATA/NOVEMBER/COMBINED/experiment_organized.yaml'

with open(yaml_file, 'w') as f:
    f.write('Epoch:\n')
    for exp in experiments:
        f.write(f'{exp.timestamp_file}\n')
```
