```python
# Use experiment_organized.yaml to find all trial IDs and put in a list
import matplotlib.pyplot as plt
from experimentStruct import ExperimentStruct
exp_yaml = r'/Users/nick/Projects/cheeseboardAnalysis/DATA/MARCH/experiment_organized.yaml'
board_file = r'/Users/nick/Projects/cheeseboardAnalysis/DATA/MARCH/board_calibration.csv'

experiments = []
with open(exp_yaml, 'r') as file:
    # Print each line of the YAML file
    for line in file:
        if '.csv' in line:
            # Remove the '- ' from the beginning of the line
            str_line = line.replace('- ', '').strip()
            experiments.append(str_line)

print(len(experiments))
```

```python
# Save all plots to a folder specified
save_folder = r'/Users/nick/Projects/cheeseboardAnalysis/FIGURES/March25/'
for exp in experiments:
    trial_data = ExperimentStruct(exp, board_file)
    trial_data.plot_all_trials()
    save_name = trial_data.experimentTag + '_Trials.png'
    plt.savefig(save_folder + save_name)
```
